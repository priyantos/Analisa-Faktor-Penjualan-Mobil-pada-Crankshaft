## Mempelajari dan Menangani Outlier
Berdasarkan hasil sebelumnya, kita akan tentukan kolom apa yang mungkin berisi outliers dan membuat sebuah DataFrame tanpa outliers tersebut. Outliers bisa diidentifikasi dengan menentukan batas bawah/atas untuk rentang nilai normal,

- Mari kita tentukan outliers pada kolom yang kita tentukan kita akan memilih

<div class="alert alert-info">
<b>Chamdani's info v.1</b> <a class="tocSkip"></a>

Outlier merupakan nilai data yang tak biasa dari suatu kolom. Dia terkadang bisa lebih kecil atau lebih besar dari kecenderungan nilai dari kolom tersebut. Lihatlah histogram yang kamu buat diatas, range nilai pada sumbu x terlihat sangat jauh. Hal ini memperlihatkan bahwa kemungkinan percilan atau outlier berada pada rentang nilai diatas nilai Q3 dan upper limit dari boxplot. Dan itu terbukti dalam boxplot yang kamu tampilkan diatas. 
    
**Perhatikan gambaran berikut**


</div>

````python
def upper_lower_whisker(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_whisker = Q1 - (1.5 * IQR)
    upper_whisker = Q3 + (1.5 * IQR)
    return lower_whisker, upper_whisker
````

<div class="alert alert-success">
<b>Chamdani's comment v.1</b> <a class="tocSkip"></a>

Hebat, Kerja bagus!

</div>

- Untuk data yang tidak akan kita gunakan yaitu cylinders dan condition, karena boxplot kurang cocok untuk tipe data kategori bersifat bertingkat.

````python
def outlier_cleansing(dataframe, column):

    lower, upper = upper_lower_whisker(dataframe,column)
    print(f'lower values : {lower}')
    print(f'upper values : {upper}')

    plt.title('Before cleaning outlier!')
    sns.boxplot(data=dataframe, x=column)
    plt.show()

    dataframe = dataframe[(dataframe[column] > lower) & (dataframe[column] < upper)]

    lower_check, upper_check = upper_lower_whisker(dataframe,column)
    print(f'lower_check values : {lower_check}')
    print(f'upper_check values : {upper_check}')

    plt.title('After cleaning outlier!')
    sns.boxplot(data=dataframe, x=column)
    plt.show()

    if lower == lower_check or upper == upper_check:
        print(f'\n====== [FINISHED] outlier in [{column}] column has been clear! ====== \n')
        return dataframe

    else:
        print(f'\n------ [RE-START] cleaning outlier in {column} column ------ \n')
        dataframe = outlier_cleansing(dataframe=dataframe, column=column)
        return dataframe
````

<div class="alert alert-danger">
